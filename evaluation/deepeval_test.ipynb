```python
from deepeval.metrics import HallucinationMetric
from deepeval.test_case import LLMTestCase
from deepeval import evaluate
import pandas as pd
from rich import print
from dotenv import load_dotenv
load_dotenv("../.env")
```

```python
metric = HallucinationMetric()

df = pd.read_csv("graphrag_run_dataset.csv")
rag_df = pd.read_csv("rag_run_dataset.csv")
```

<h3>GRAPHRAG</h3>

```python
test_cases = []

for _, row in df.iterrows():

    tc = LLMTestCase(
        input=row["question"],
        actual_output=row["generated_answer"],
        expected_output=row["expected_answer"],
        context=[row["retrieved_context"]],
    )

    test_cases.append(tc)
```

```python
results = evaluate(
    test_cases=test_cases,
    metrics=[metric],
)
```

```python
results = evaluate(
    test_cases=test_cases,
    metrics=[metric],
)
```

```python
for i in range(len(results.test_results)):
    # i = 3
    # print(results.test_results[i].__dict__.keys())
    res = results.test_results[i]
    total_metrics = len(res.metrics_data)
    print(f"Name: {res.name} | Index: {res.index} | #Metrics: {total_metrics}\n\nQuestion: {res.input}")
    # print(results.test_results[0])
    # print(results.test_results[i].metrics_data[j].__dict__.keys())
    # print((results.test_results[i].metrics_data))
    
    for n_metric in range(total_metrics):
        met = res.metrics_data[n_metric]
        print(f"Metric: {met.name} | Threshold: {met.threshold} | Success: {met.success} | Score: {met.score} | Cost: ${met.evaluation_cost:.2f}\n\nReason: {met.reason}")
        # print(f"Verdict: {met.verbose_logs.strip()}")
        # print(f"Verdict: {repr(met.verbose_logs.strip())}")
        # print(f"Verdict: {dict(met.verbose_logs.strip())}")
```
